This notebook is primarily used for drawing BER VS SNR graph for autoencoder experiments.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns

plt.rcParams.update({'font.size': 18})
```

```python
scheme_1_mode_concat_k_4_n_1_no_pred = pd.read_csv("autoencoder/no_pred_net_downlink_downlink_16_xant_one_hot_seed_42_version_2_mode_concat_k_4_n_1/bler_ber.csv")
scheme_1_mode_concat_k_4_n_2_no_pred = pd.read_csv("autoencoder/no_pred_net_downlink_downlink_16_xant_one_hot_seed_42_version_2_mode_concat_k_4_n_2/bler_ber.csv")
scheme_1_mode_concat_k_4_n_4_no_pred = pd.read_csv("autoencoder/no_pred_net_downlink_downlink_16_xant_one_hot_seed_42_version_2_mode_concat_k_4_n_4/bler_ber.csv")

scheme_1_mode_concat_k_4_n_1_pred = pd.read_csv("autoencoder/pred_net_downlink_downlink_16_xant_one_hot_seed_42_version_2_mode_concat_k_4_n_1/bler_ber.csv")
scheme_1_mode_concat_k_4_n_2_pred = pd.read_csv("autoencoder/pred_net_downlink_downlink_16_xant_one_hot_seed_42_version_2_mode_concat_k_4_n_2/bler_ber.csv")
scheme_1_mode_concat_k_4_n_4_pred = pd.read_csv("autoencoder/pred_net_downlink_downlink_16_xant_one_hot_seed_42_version_2_mode_concat_k_4_n_4/bler_ber.csv")
```

```python
scheme_1_mode_dot_k_4_n_1_no_pred = pd.read_csv("autoencoder/no_pred_net_downlink_downlink_16_xant_one_hot_seed_42_version_2_mode_dot_k_4_n_1/bler_ber.csv")
scheme_1_mode_dot_k_4_n_2_no_pred = pd.read_csv("autoencoder/no_pred_net_downlink_downlink_16_xant_one_hot_seed_42_version_2_mode_dot_k_4_n_2/bler_ber.csv")
scheme_1_mode_dot_k_4_n_4_no_pred = pd.read_csv("autoencoder/no_pred_net_downlink_downlink_16_xant_one_hot_seed_42_version_2_mode_dot_k_4_n_4/bler_ber.csv")

scheme_1_mode_dot_k_4_n_1_pred = pd.read_csv("autoencoder/pred_net_downlink_downlink_16_xant_one_hot_seed_42_version_2_mode_dot_k_4_n_1/bler_ber.csv")
scheme_1_mode_dot_k_4_n_2_pred = pd.read_csv("autoencoder/pred_net_downlink_downlink_16_xant_one_hot_seed_42_version_2_mode_dot_k_4_n_2/bler_ber.csv")
scheme_1_mode_dot_k_4_n_4_pred = pd.read_csv("autoencoder/pred_net_downlink_downlink_16_xant_one_hot_seed_42_version_2_mode_dot_k_4_n_4/bler_ber.csv")
```

```python
scheme_5_mode_concat_k_4_n_1 = pd.read_csv("autoencoder/uplink+feats_downlink_16_xant_one_hot_seed_42_version_2_mode_concat_k_4_n_1/bler_ber.csv")
scheme_5_mode_concat_k_4_n_2 = pd.read_csv("autoencoder/uplink+feats_downlink_16_xant_one_hot_seed_42_version_2_mode_concat_k_4_n_2/bler_ber.csv")
scheme_5_mode_concat_k_4_n_4 = pd.read_csv("autoencoder/uplink+feats_downlink_16_xant_one_hot_seed_42_version_2_mode_concat_k_4_n_4/bler_ber.csv")
```

```python
scheme_5_mode_dot_k_4_n_1 = pd.read_csv("autoencoder/uplink+feats_downlink_16_xant_one_hot_seed_42_version_2_mode_dot_k_4_n_1/bler_ber.csv")
scheme_5_mode_dot_k_4_n_2 = pd.read_csv("autoencoder/uplink+feats_downlink_16_xant_one_hot_seed_42_version_2_mode_dot_k_4_n_2/bler_ber.csv")
scheme_5_mode_dot_k_4_n_4 = pd.read_csv("autoencoder/uplink+feats_downlink_16_xant_one_hot_seed_42_version_2_mode_dot_k_4_n_4/bler_ber.csv")
```

```python
def plot_single_bler(df, legend):
    # plot till 5 db
    plt.plot(df['snr'], np.round(df['bler'], 3))
    plt.scatter(df['snr'], np.round(df['bler'], 3), label=legend)

plt.figure(figsize=(10,8))

# scheme 1 concat mode
plot_single_bler(scheme_1_mode_concat_k_4_n_4_no_pred, 'scheme_1_mode_concat_n_4_no_pred')
plot_single_bler(scheme_1_mode_concat_k_4_n_4_pred, 'scheme_1_mode_concat_n_4_pred')

# scheme 1 dot mode
plot_single_bler(scheme_1_mode_dot_k_4_n_4_no_pred, 'scheme_1_mode_dot_n_4_no_pred')
plot_single_bler(scheme_1_mode_dot_k_4_n_4_pred, 'scheme_1_mode_dot_n_4_pred')

# scheme 5 concat mode
plot_single_bler(scheme_5_mode_concat_k_4_n_4, 'scheme_5_mode_concat_k_4_n_4')

# scheme 5 dot mode
plot_single_bler(scheme_5_mode_dot_k_4_n_4, 'scheme_5_mode_dot_k_4_n_4')

plt.yscale('log')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.title("BLER vs snr for different schemes")
plt.xlabel("Eb/N0(dB)")
plt.ylabel("BLER")
plt.show()
```
