This notebook is primarily used for drawing BER VS SNR graph for autoencoder experiments.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns

plt.rcParams.update({'font.size': 20})
```

## 128 ANTENNA

```python
dl_dl_128_xant = pd.read_csv("autoencoder/downlink_downlink_128_xant.csv")
estdl_dl_128_xant = pd.read_csv("autoencoder/estimateddownlink_downlink_128_xant.csv")
estdlfeats_dl_128_xant = pd.read_csv("autoencoder/estimateddownlink+feats_downlink_128_xant.csv")
ul_dl_128_xant = pd.read_csv("autoencoder/uplink_downlink_128_xant.csv")
ulfeats_dl_128_xant = pd.read_csv("autoencoder/uplink+feats_downlink_128_xant.csv")
```

```python
plt.figure(figsize=(10,8))
plt.plot(dl_dl_128_xant['snr'].iloc[:-15], np.round(dl_dl_128_xant['bler'].iloc[:-15], 3))
plt.scatter(dl_dl_128_xant['snr'].iloc[:-15], np.round(dl_dl_128_xant['bler'].iloc[:-15], 3), label='dl-dl')

plt.plot(estdl_dl_128_xant['snr'].iloc[:-15], np.round(estdl_dl_128_xant['bler'].iloc[:-15], 3))
plt.scatter(estdl_dl_128_xant['snr'].iloc[:-15], np.round(estdl_dl_128_xant['bler'].iloc[:-15], 3),label='estdl-dl')

plt.plot(estdlfeats_dl_128_xant['snr'].iloc[:-15], np.round(estdlfeats_dl_128_xant['bler'].iloc[:-15],3))
plt.scatter(estdlfeats_dl_128_xant['snr'].iloc[:-15], np.round(estdlfeats_dl_128_xant['bler'].iloc[:-15],3), label='estdl+feats-dl')

plt.plot(ulfeats_dl_128_xant['snr'].iloc[:-15], np.round(ulfeats_dl_128_xant['bler'].iloc[:-15], 3))
plt.scatter(ulfeats_dl_128_xant['snr'].iloc[:-15], np.round(ulfeats_dl_128_xant['bler'].iloc[:-15], 3), label='ul+feats-dl')

plt.plot(ul_dl_128_xant['snr'].iloc[:-15], np.round(ul_dl_128_xant['bler'].iloc[:-15], 3))
plt.scatter(ul_dl_128_xant['snr'].iloc[:-15], np.round(ul_dl_128_xant['bler'].iloc[:-15], 3), label='ul-dl')

plt.yscale('log')
plt.legend()
plt.title("BLER vs snr for different schemes")
plt.xlabel("Eb/N0(dB)")
plt.ylabel("BLER")
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.plot(dl_dl_128_xant['snr'].iloc[:-15], np.round(dl_dl_128_xant['ber'].iloc[:-15], 3))
plt.scatter(dl_dl_128_xant['snr'].iloc[:-15], np.round(dl_dl_128_xant['ber'].iloc[:-15], 3), label='dl-dl')

plt.plot(estdl_dl_128_xant['snr'].iloc[:-15], np.round(estdl_dl_128_xant['ber'].iloc[:-15], 3))
plt.scatter(estdl_dl_128_xant['snr'].iloc[:-15], np.round(estdl_dl_128_xant['ber'].iloc[:-15], 3),label='estdl-dl')

plt.plot(estdlfeats_dl_128_xant['snr'].iloc[:-15], np.round(estdlfeats_dl_128_xant['ber'].iloc[:-15],3))
plt.scatter(estdlfeats_dl_128_xant['snr'].iloc[:-15], np.round(estdlfeats_dl_128_xant['ber'].iloc[:-15],3), label='estdl+feats-dl')

plt.plot(ulfeats_dl_128_xant['snr'].iloc[:-15], np.round(ulfeats_dl_128_xant['ber'].iloc[:-15], 3))
plt.scatter(ulfeats_dl_128_xant['snr'].iloc[:-15], np.round(ulfeats_dl_128_xant['ber'].iloc[:-15], 3), label='ul+feats-dl')

plt.plot(ul_dl_128_xant['snr'].iloc[:-15], np.round(ul_dl_128_xant['ber'].iloc[:-15], 3))
plt.scatter(ul_dl_128_xant['snr'].iloc[:-15], np.round(ul_dl_128_xant['ber'].iloc[:-15], 3), label='ul-dl')

plt.yscale('log')
plt.legend()
plt.title("BER vs snr for different schemes")
plt.xlabel("Eb/N0(dB)")
plt.ylabel("BER")
plt.show()
```
